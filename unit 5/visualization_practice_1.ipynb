```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Manually creating the 'tips' dataset for demonstration
data = {
    "total_bill": [16.99, 10.34, 21.01, 23.68, 24.59, 25.29, 8.77, 26.88, 15.04, 14.78],
    "tip": [1.01, 1.66, 3.50, 3.31, 3.61, 4.71, 2.00, 3.12, 1.96, 3.23],
    "sex": ["Female", "Male", "Male", "Male", "Female", "Male", "Male", "Male", "Female", "Male"],
    "smoker": ["No", "No", "No", "No", "Yes", "No", "Yes", "No", "Yes", "Yes"],
    "day": ["Sun", "Sun", "Sun", "Sun", "Sun", "Sun", "Sun", "Sun", "Sun", "Sun"],
    "time": ["Dinner", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner"],
    "size": [2, 3, 3, 2, 4, 4, 2, 4, 2, 2],
}

# Creating a DataFrame
df = pd.DataFrame(data)

# 1. Bar plot to display the count of occurrences of a categorical variable ('sex')
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='sex', palette='pastel')
plt.title('Count of Occurrences by Sex', fontsize=16)
plt.xlabel('Sex', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

# 2. Histogram to visualize the distribution of a numerical variable ('total_bill')
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='total_bill', kde=True, color='blue')
plt.title('Distribution of Total Bill', fontsize=16)
plt.xlabel('Total Bill', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()

# 3. Box plot to compare the distribution of 'total_bill' across 'sex'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='sex', y='total_bill', palette='coolwarm')
plt.title('Box Plot of Total Bill by Sex', fontsize=16)
plt.xlabel('Sex', fontsize=12)
plt.ylabel('Total Bill', fontsize=12)
plt.show()

# 4. Scatter plot to visualize the relationship between 'total_bill' and 'tip'
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex', palette='viridis', s=100)
plt.title('Scatter Plot of Total Bill vs Tip', fontsize=16)
plt.xlabel('Total Bill', fontsize=12)
plt.ylabel('Tip', fontsize=12)
plt.legend(title='Sex')
plt.show()

# 5. Line plot to show the trend of 'total_bill' over a continuous range (sorted by index)
plt.figure(figsize=(8, 6))
plt.plot(df['total_bill'].sort_values().reset_index(drop=True), color='green', marker='o', linestyle='dashed')
plt.title('Trend of Total Bill', fontsize=16)
plt.xlabel('Index (sorted)', fontsize=12)
plt.ylabel('Total Bill', fontsize=12)
plt.show()

# 6. Pie chart to visualize the percentage distribution of 'smoker'
plt.figure(figsize=(8, 6))
df['smoker'].value_counts().plot.pie(autopct='%1.1f%%', colors=['gold', 'skyblue'], labels=['No', 'Yes'])
plt.title('Percentage Distribution of Smokers', fontsize=16)
plt.ylabel('')  # Remove y-label for clarity
plt.show()

# 7. Heatmap to display the correlation matrix of numerical variables
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Variables', fontsize=16)
plt.show()

# 8. Customization: Modifying a previous plot (Scatter Plot)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex', palette='Set1', s=120, edgecolor='black')
plt.title('Customized Scatter Plot of Total Bill vs Tip', fontsize=16)
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Tip ($)', fontsize=12)
plt.legend(title='Gender', fontsize=10)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()

# 9. Histogram with modified bins and colors
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='total_bill', bins=5, kde=True, color='purple')
plt.title('Modified Histogram of Total Bill', fontsize=16)
plt.xlabel('Total Bill', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()

# 10. A second heatmap with more emphasis on styling
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='YlGnBu', linewidths=0.5, linecolor='black', fmt='.2f')
plt.title('Styled Correlation Matrix', fontsize=16)
plt.show()
```
