### Univariate Distributions

1. Use Seaborn to plot a histogram of the tips datasetâ€™s total_bill column. Customize the histogram by setting bins=15 and color as skyblue.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.histplot(tips["total_bill"], bins=15, color="skyblue")
plt.title("Histogram of Total Bill")
plt.show()
```

2. Plot the KDE for the total_bill column in the tips dataset using Seaborn. Add a title "Total Bill Distribution" and set the line color to green.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.kdeplot(tips["total_bill"], color="green")
plt.title("Total Bill Distribution")
plt.show()
```

3. Create an overlay plot of the histogram and KDE for the total_bill column from the tips dataset. Set the histogram color to light gray with bins=20 and the KDE color to purple.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.histplot(tips["total_bill"], bins=20, color="lightgray", kde=True)
sns.kdeplot(tips["total_bill"], color="purple", linewidth=2)
plt.title("Histogram and KDE Overlay for Total Bill")
plt.show()
```

4. Plot a rug plot for the tip column in the tips dataset using Seaborn. Overlay a KDE on top and set the color for the KDE to orange and line width to 2.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.kdeplot(tips["tip"], color="orange", linewidth=2)
sns.rugplot(tips["tip"], color="orange")
plt.title("Rug Plot and KDE for Tips")
plt.show()
```

5. Use Seaborn to create a box plot of the total_bill column from the tips dataset. Set the color of the box to teal and display the plot vertically.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.boxplot(y=tips["total_bill"], color="teal")
plt.title("Box Plot of Total Bill")
plt.show()
```

### Bivariate Distribution

1. Use Seaborn to plot a scatter plot of total_bill vs. tip from the tips dataset. Add a regression line and customize the scatter plot points as blue circles with a size of 50.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.regplot(x="total_bill", y="tip", data=tips, scatter_kws={"color": "blue", "s": 50})
plt.title("Scatter Plot of Total Bill vs Tip with Regression Line")
plt.show()
```

2. Plot a joint plot of total_bill and tip using Seabornâ€™s jointplot() function from the tips dataset. Display both scatter and histogram plots, with the scatter plot color set to coral.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter", color="coral")
plt.suptitle("Joint Plot of Total Bill vs Tip (Scatter & Histogram)", y=1.02)
plt.show()
```

3. Create a hexbin plot of the total_bill and tip columns in the tips dataset using Seaborn. Use a joint plot with kind='hex' and set the color map to coolwarm.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex", cmap="coolwarm")
plt.suptitle("Hexbin Plot of Total Bill vs Tip", y=1.02)
plt.show()
```
