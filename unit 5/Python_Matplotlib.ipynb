### Basic Plotting
1. Write a Python program using Matplotlib to plot a simple line graph with x = [0, 1, 2, 3, 4] and y = [0, 1, 4, 9, 16]. Customize the line with:
A green color
A dashed line style
Circle markers at each data point.

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.plot(x, y, color='green', linestyle='--', marker='o')
plt.show()
```

2. Plot a simple line graph with x = [1, 2, 3, 4, 5] and y = [10, 20, 25, 30, 35]. Add:
A title "Growth over Time"
X-axis label as "Time"
Y-axis label as "Growth"
Customize the line with a blue color and a marker size of 8.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

plt.plot(x, y, color='blue', marker='o', markersize=8)
plt.title("Growth over Time")
plt.xlabel("Time")
plt.ylabel("Growth")
plt.show()
```

3. Write a Python program to plot two line graphs on the same plot using Matplotlib:
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 9] (with a red solid line)
y2 = [1, 4, 6, 8, 10] (with a blue dashed line)
Add a legend to distinguish the two lines.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 9]
y2 = [1, 4, 6, 8, 10]

plt.plot(x, y1, color='red', label="Line 1", linestyle='-')
plt.plot(x, y2, color='blue', label="Line 2", linestyle='--')
plt.legend()
plt.show()
```

### Subplots

4. Creating Subplots with Different Sizes

Write a Python program to create two subplots in a 1x2 grid:
The first subplot should display a simple line graph with x = [1, 2, 3, 4, 5] and y = [5, 6, 7, 8, 9].
The second subplot should display the same data as a scatter plot, using a green color for the markers and adjusting the marker size to 40.
Use Matplotlib's figsize argument to set the figure size to (12, 4).

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 6, 7, 8, 9]

plt.figure(figsize=(12, 4))

# First subplot
plt.subplot(1, 2, 1)
plt.plot(x, y)

# Second subplot
plt.subplot(1, 2, 2)
plt.scatter(x, y, color='green', s=40)

plt.show()
```

5. Create a 2x1 subplot where:
The top subplot is a histogram for the data [5, 6, 8, 8, 10, 10, 10, 12, 14, 16, 18, 18, 20], with bins set to 5 and a blue color.
The bottom subplot is a line plot of x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10] with a red line and circle markers.

```python
import matplotlib.pyplot as plt

data = [5, 6, 8, 8, 10, 10, 10, 12, 14, 16, 18, 18, 20]
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(8, 6))

# Top subplot (Histogram)
plt.subplot(2, 1, 1)
plt.hist(data, bins=5, color='blue')
plt.title("Histogram")

# Bottom subplot (Line plot)
plt.subplot(2, 1, 2)
plt.plot(x, y, color='red', marker='o')
plt.title("Line Plot")

plt.tight_layout()
plt.show()
```

8. Create a box plot for the data data1 = [10, 15, 14, 20, 25, 30, 35, 40, 45] and data2 = [20, 25, 30, 35, 40, 45, 50, 55, 60] in a single figure.
Display both data sets in the same box plot, with different colors for each box.

9. Write a program to plot x = [1, 2, 3, 4, 5] and y = [1, 4, 9, 16, 25] in both a line plot and a scatter plot within the same figure.
Customize:
The line plot color to blue with a dashed line style

10. Plot a simple line graph with x = [1, 2, 3, 4, 5] and y = [2, 3, 5, 7, 11].
Add:
A grid to the plot
X-axis limits from 0 to 6
Y-axis limits from 0 to 12
Set the line to a green color with a line width of 2

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, color='green', linewidth=2)
plt.grid()
plt.xlim(0, 6)
plt.ylim(0, 12)
plt.show()
```
